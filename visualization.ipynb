```python
import json
import matplotlib.pyplot as plt
```

```python
def plot_graphs(title, baseline, adaact):

    with open(baseline, 'r') as file:
        baseline_data = json.load(file)

    with open(adaact, 'r') as file:
        adaact_data = json.load(file)

    baseline_train_loss = baseline_data['train_loss']
    baseline_train_accuracy = baseline_data['train_accuracy']
    baseline_test_loss = baseline_data['test_loss']
    baseline_test_accuracy = baseline_data['test_accuracy']

    adaact_train_loss = adaact_data['train_loss']
    adaact_train_accuracy = adaact_data['train_accuracy']
    adaact_test_loss = adaact_data['test_loss']
    adaact_test_accuracy = adaact_data['test_accuracy']

    epochs =[i for i in range(1, len(baseline_train_loss)+1)]
    
    #plotting the training loss
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    fig.suptitle(title, fontsize=16)

    axs[0, 0].plot(epochs, baseline_train_loss, color='m', label="Baseline")
    axs[0, 0].plot(epochs, adaact_train_loss, color='g', label="With AdaAct")
    axs[0, 0].set_title("Training Loss")
    axs[0, 0].set_xlabel("Epochs")
    axs[0, 0].set_label("Loss")
    axs[0, 0].grid(True)
    axs[0, 0].legend()
    
    #plotting the training accuracy
    axs[0, 1].plot(epochs, baseline_train_accuracy, color='m', label="Baseline")
    axs[0, 1].plot(epochs, adaact_train_accuracy, color='g', label="With AdaAct")
    axs[0, 1].set_title("Training Accuracy")
    axs[0, 1].set_xlabel("Epochs")
    axs[0, 1].set_label("Accuracy")
    axs[0, 1].grid(True)
    axs[0, 1].legend()
    
    #plotting the testing loss
    axs[1, 0].plot(epochs, baseline_test_loss, color='m', label="Baseline")
    axs[1, 0].plot(epochs, adaact_test_loss, color='g', label="With AdaAct")
    axs[1, 0].set_title("Testing Loss")
    axs[1, 0].set_xlabel("Epochs")
    axs[1, 0].set_label("Loss")
    axs[1, 0].grid(True)
    axs[1, 0].legend()
    
    #plotting the testing accuracy
    axs[1, 1].plot(epochs, baseline_test_accuracy, color='m', label="Baseline")
    axs[1, 1].plot(epochs, adaact_test_accuracy, color='g', label="With AdaAct")
    axs[1, 1].set_title("Testing Accuracy")
    axs[1, 1].set_xlabel("Epochs")
    axs[1, 1].set_label("Accuracy")
    axs[1, 1].grid(True)
    axs[1, 1].legend()
    
    plt.tight_layout()
    plt.show()
        
```

```python
plot_graphs('Resnet18 on Cifar10 for 100 Epochs', 
            'checkpoints/resnet18_adaactFalse_cifar10_epochs100.json', 
            'checkpoints/resnet18_adaactTrue_cifar10_epochs100.json')
```

```python
plot_graphs('Resnet18 on Cifar100 for 100 Epochs', 
            'checkpoints/resnet18_adaactFalse_cifar100_epochs100.json', 
            'checkpoints/resnet18_adaactTrue_cifar100_epochs100.json')
```
