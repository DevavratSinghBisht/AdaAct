```python
import matplotlib.pyplot as plt
```

```python
def plot_graphs(train_loss, train_accuracy, test_loss, test_accuracy):
    epochs =[]
    for i in range(1, 100+1):
        epochs.append(i)
    
    #plotting the training loss
    fig, axs = plt.subplots(2, 2, figsize=(12, 10))
    axs[0, 0].plot(epochs, train_loss, color='b', label="Training Loss")
    axs[0, 0].set_title("Training Loss")
    axs[0, 0].set_xlabel("Epochs")
    axs[0, 0].set_label("Loss")
    axs[0, 0].grid(True)
    axs[0, 0].legend()
    
    #plotting the training accuracy
    axs[0, 1].plot(epochs, train_accuracy, color='g', label="Training Accuracy")
    axs[0, 1].set_title("Training Accuracy")
    axs[0, 1].set_xlabel("Epochs")
    axs[0, 1].set_label("Accuracy")
    axs[0, 1].grid(True)
    axs[0, 1].legend()
    
    #plotting the testing loss
    axs[1, 0].plot(epochs, test_loss, color='r', label="Testing Loss")
    axs[1, 0].set_title("Testing Loss")
    axs[1, 0].set_xlabel("Epochs")
    axs[1, 0].set_label("Loss")
    axs[1, 0].grid(True)
    axs[1, 0].legend()
    
    #plotting the testing accuracy
    axs[1, 1].plot(epochs, test_accuracy, color='m', label="Testing Accuracy")
    axs[1, 1].set_title("Testing Accuracy")
    axs[1, 1].set_xlabel("Epochs")
    axs[1, 1].set_label("Accuracy")
    axs[1, 1].grid(True)
    axs[1, 1].legend()
    
    plt.tight_layout()
    plt.show()

# train_loss = [1.9728 - (i*0.01) for i in range(100)]
# train_accuracy = [87.738 + (i*0.5) for i in range(100)]
# test_loss = [0.3504 - (i*0.002) for i in range(100)]
# test_accuracy = [36.81 + (i*0.3) for i in range(100)]

if __name__ == "__main__":
    plot_graphs(train_loss, train_accuracy, test_loss, test_accuracy)
        
```
